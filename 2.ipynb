```python
def count_freq(lines,symbols):
    freq = {}
    for i in symbols: freq[i] = 0
    num = 0
    
    for l in lines:
        l = l.lower()
        l = l.strip()
        for s in symbols:
            freq[s] += l.count(s)
            num += l.count(s)
            
    for s in freq:
        freq[s] /= num
    
    freq = {k: v for k, v in sorted(freq.items(), key=lambda item: item[1])[::-1]}
        
    return freq
```

```python
def count_freq2(lines,symbols,low=True):
    freq = {}
    for i in symbols: freq[i] = 0
    num = 0
    
    for l in lines:
        if low: l = l.lower()
        for i in range(0,len(l),2):
            if l[i:i+2] in symbols:
                freq[l[i:i+2]] += 1
                num += 1
                  
    for s in freq:
        freq[s] /= num
    
    freq = {k: v for k, v in sorted(freq.items(), key=lambda item: item[1])[::-1]}
        
    return freq
```

# Применим MCMC-сэмплирование

#### Распределение в пространстве перестановок будет выглядеть так:
#### p( количество транспозиций = i ) * ∏p( переставлены элементы x1 и x2 для транспозиции k)

```python
def my_distribution1(l,probs1):
    sample = np.random.choice(np.append(np.arange(l),[0]),p = probs1)
    return sample
```

```python
def my_distribution2(ar,probs2,num_transpose=0):
    ar = np.array(ar)
    transpositions = [np.where(ar == i)[0][0] for i in np.random.choice(ar,num_transpose*2,False,p = probs2)]
    transpositions = np.resize(transpositions,(num_transpose,2))
    return transpositions
```

```python
def find_probs_for_dists(p,l,ar,ar_probs):
    probs1 = (1/p-1)*p**(np.arange(l)+1)
    probs1 = np.append(probs1,1-np.sum(probs1))
    
#     probs2 = ar_probs
    
    q = sum([1/i for i in ar_probs if i != 0])
    probs2 = [(1/i)/q for i in ar_probs if i != 0]
    probs2 += [0]*ar_probs.count(0)
    
    #probs2[0] += 1-sum(probs2)
    return probs1,probs2
```

```python
def my_MCMC(orig,orig_probs,num_sampl,p=1/2,l=10,step=100,walk=10):
    orig = np.array(orig)
    c = np.arange(len(orig))
    probs1,probs2 = find_probs_for_dists(p,l,orig,orig_probs)
    res = []
    
    #сначала погуляем немного, потом посэмплируем
    for i in range(num_sampl+walk):
        for j in range(step):
        
            num_transpose = my_distribution1(l,probs1)
            distance1 = len([i for i in range(len(c)) if c[i] != i])
            c_c = c
            
            if num_transpose == 0:
                continue
                
            where = 1
            transpositions = my_distribution2(orig,probs2,num_transpose)
            for t in transpositions:
                c_c[t] = c_c[[t[1],t[0]]]
                where *= 2*(1-orig_probs[t[0]])*(1-orig_probs[t[0]])
                
            distance2 = len([i for i in range(len(c)) if c_c[i] != i])
            
            how_many = ((1/p-1)*p**(distance2+1))/((1/p-1)*p**(distance1+1))
            a = how_many*where
            x = np.random.random()
            if x < a:
                c = c_c
        if i >= walk:
            res += [orig[c]]
    
    return res
```

```python
k = 0
for l in cdt[16:18]:
    for i in range(0,len(l),2):
        k += 1
```

```python
cd_freq2_perms = my_MCMC(list(cd_freq2.keys()),list(cd_freq2.values()),k,1/2,5,100,10)
```

```python
ln1 = ''
for l in cdt[16:18]:
    for i in range(0,len(l),2):
        if l[i:i+2].lower() in letters2:
            ln1 += list(wp_freq2.keys())[list(cd_freq2.keys()).index(l[i:i+2].lower())]
        else: ln1 += l[i:i+2]
    ln1 += '\n'
            
ln2 = ''
for l in cdt[16:18]:
    k1 = 0
    for i in range(0,len(l),2):
        if l[i:i+2].lower() in letters2:
            ln2 += list(wp_freq2.keys())[list(cd_freq2_perms[k1]).index(l[i:i+2].lower())]
            k1 += 1
        else: 
            ln2 += l[i:i+2]
    ln2 += '\n'
```
